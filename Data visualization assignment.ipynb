```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
pca = decomposition.PCA(n_components=3).fit_transform(X)
fig = plt.figure(1, figsize=(10, 7))
ax = Axes3D(fig, elev=-130, azim=100)
ax.scatter(pca[:, 0], pca[:, 1], pca[:, 2], c=y, cmap=plt.cm.Set1, edgecolor='g', s=45)
ax.set_title("PCA AXIS")
ax.set_xlabel("1st PCA")
ax.set_ylabel("2nd PCA")
ax.set_zlabel("3rd PCA")
plt.show()
```
